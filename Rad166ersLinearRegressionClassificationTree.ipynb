## Multiple Regression Model

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv("/Life Expectancy Data.csv")
df.dropna(inplace=True) # Dropping all values with NA data to provide more accurate data for Regression model

x = df[['Life expectancy ']]
y = df[['Alcohol']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=69)

model = LinearRegression()

model.fit(x_train, y_train)

y_pred = model.predict(x_test) # Prediction of y from linear regression model
```

We followed the same steps so far from the Simple Regression. Now, we are going to evaluate the true versus the predicted values for this set of data:

```python
plt.scatter(y_test, y_pred, color='black')
plt.title('True vs. Predicted values of Test set')
plt.xlabel('True y values')
plt.ylabel('Predicted y values')

# model eval
print("Root mean squared error = %.5f" % np.sqrt(mean_squared_error(y_test, y_pred)))
print('R^2 = %.5f' % r2_score(y_test, y_pred))
```

```python
# plot
plt.scatter(x_test, y_test, color='black')
plt.plot(x_test, y_pred, color='red')
titleString = 'Predicted Function: y = %.2fx + %.2f' % (model.coef_.item(), model.intercept_.item())
plt.title(titleString)
plt.xlabel('Life Expectancy (Years)')
plt.ylabel('Alchohol per capita (consumption in litres of pure alcohol)')

# line plot exact numbers
print('Slope = ', model.coef_.item())
print('Intercept = ', model.intercept_.item())
```

We want to discretize the column "Chance of Admit"

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import os

data = pd.read_csv(r"/content/sample_data/Admission_Predict_Ver1.1_small_data_set_for_Linear_Regression.csv")

data.head()
```

Above is example of the data and we only want to discretize the last column which is "Chance of Admit" into 3 classes so being percentages what we thought was good to use for the 3 classes was something like High Chance > 75, Medium Chance between 74 and and 40 and Low Chance between 39 and 0.

Using the pandas cut function we can use only the Chance of Admit column and put it into 3 classes and assign back to the dataframe

```python
# Use pd.cut to discretize "Chance of Admit" into the three categories
data['Classes'] = pd.cut(
    data['Chance of Admit'],  # Column to discretize
    bins=3,  # Bin edges
    labels=['Low', 'Medium', 'High'],  # Labels for the bins
    include_lowest=True  # To include the lowest value in the first bin
)

# Display the first few rows to check the new categorization
data[['Chance of Admit', 'Classes']].head()
```

Now we need to split the data set into testing and training sets

```python
x = data.drop(['Chance of Admit', "Classes"], axis=1)  # Features
y = data['Classes']  # Target variable

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training 30% testing
```

Now we can build the decision tree model and evaluate it

```python
# Create a DecisionTreeClassifier object
clf = DecisionTreeClassifier()

# Fit the classifier to the training data
clf.fit(x_train, y_train)

# Predict the classes of the testing data
y_pred = clf.predict(x_test)

# Evaluate the performance of the classifier
accuracy = accuracy_score(y_test, y_pred)

# Accuracy denotes the percentage of correct predictions
print(f"Accuracy: {accuracy:.2f}")
```

Here is a visualization of the classification tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))  # Set figure size for better readability
plot_tree(clf, filled=True, rounded=True,
          class_names=['Low', 'Medium', 'High'],
          feature_names=x.columns.tolist())  # Convert Index to list
plt.show()
```

Since the tree above is unreadable I'm going to set the max depth of the tree to be 3 so we can have a clearer look at the tree


```python
clf = DecisionTreeClassifier(max_depth=3) # Set the max depth to 3

# Fit the classifier to the training data
clf.fit(x_train, y_train)

# Predict the classes of the testing data
y_pred = clf.predict(x_test)

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))  # Set figure size for better readability
plot_tree(clf, filled=True, rounded=True,
          class_names=['Low', 'Medium', 'High'],
          feature_names=x.columns.tolist())  # Convert Index to list
plt.show()
```

```python
# Part C.
import pandas as pd
dataset={'color':['red','blue','red','green','red','green'],
            'shape':['square','square','round','square','round','round'],
            'size':['big','big','small','small','big','big'],
            'class':['+','+','-','-','+','-']}
dataframe = pd.DataFrame(dataset)
print(dataframe)
```

```python
# Calculate initial entropy.
classEntries = dataframe['class'].count()
print('Class entries:',classEntries)

# Initial entropy variable counts.
def colVarCount(dframe, column, variable): # Get the variable count of a column
    if dframe[dframe[column]==variable].empty:
        return 0
    else:
        return dframe[column].value_counts()[variable]


classCntPos = colVarCount(dataframe, 'class', '+')
print('Number of positives:',classCntPos)

classCntNeg = colVarCount(dataframe, 'class', '-')
print('Number of negatives:',classCntNeg)


def proportion(varCnt, colEntryCnt): # What are the proportions of the initial entropy?
    return varCnt / colEntryCnt

classPptnPos = proportion(classCntPos, classEntries)
classPptnNeg = proportion(classCntNeg, classEntries)


import math # Needed for getting (base-2) logarithms.
def pValue(varProportion): # Get the p-value.
    if varProportion==0:
        return 0
    else:
        return (varProportion * (math.log2(varProportion)))

classPvPos = pValue(classPptnPos)
print('Positive\'s p-value:',classPptnPos)

classPvNeg = pValue(classPptnNeg)
print('Negative\'s p-value:',classPptnNeg)


def entropy(list):
    dst = 0
    for i in list:
        dst += (-i)
    return dst

classPvals = [classPvPos, classPvNeg]
entropyInitial = entropy(classPvals)
print('Initial entropy:',entropyInitial)
```

```python
# Calculate entropy & gain for color, shape, and size with respect to signage.
```

```python
def getEntropy(dframe, column, variable):
    df = dframe[dframe[column]==variable]
    cntPos = colVarCount(df,'class','+') # Get the number of positive and negatives.
    cntNeg = colVarCount(df,'class','-')
    entries = df['class'].count() # Get the number of entries.
    pptnPos = proportion(cntPos, entries) # Get the proportions.
    pptnNeg = proportion(cntNeg, entries)
    pvalPos = pValue(pptnPos) # Get the p-values.
    pvalNeg = pValue(pptnNeg)
    pvalList = [pvalPos, pvalNeg] # Assort the p-values into a list.
    return entropy(pvalList)

def getProportionedEntropy(column, variable, entropyVal): # Get the proportioned entropy for average entropy.
    columnEntries = dataframe[column].count()
    varCnt = colVarCount(dataframe, column, variable)
    varPptn = proportion(varCnt, columnEntries)
    return entropyVal * varPptn

def getAverageEntropy(column):
    df = set(dataframe[column]) # Only iterate through values once.
    proportionedEntropies = [] # List to hold the proportioned entropies.
    for x in df:
        xEntropy = getEntropy(dataframe, column, x)
        xProportionedEntropy = getProportionedEntropy(column, x, xEntropy)
        proportionedEntropies.append(xProportionedEntropy)
        print(x,':\n\tEntropy:',xEntropy,'\n\tProportioned Entropy:',xProportionedEntropy)
    averageEntropyVal = 0;
    for x in proportionedEntropies: # Add the proportioned entropies together for average entropy.
        averageEntropyVal += x
    return averageEntropyVal

def getGain(averageEntropy):
    return 1 - averageEntropy
```

```python
print('Entropies of Color:')
avgColorEntropy = getAverageEntropy('color')
print('\nAverage entropy of Color:',avgColorEntropy,'\nGain of Color:',getGain(avgColorEntropy))
```

```python
print('Entropies of Shape:')
avgShapeEntropy = getAverageEntropy('shape')
print('\nAverage entropy of Shape:',avgShapeEntropy,'\nGain of Shape:',getGain(avgShapeEntropy))
```

```python
print('Entropies of Size:')
avgSizeEntropy = getAverageEntropy('size')
print('\nAverage entropy of Size:',avgSizeEntropy,'\nGain of Size:',getGain(avgSizeEntropy))
```
